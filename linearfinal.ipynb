```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('/home/joseph/Downloads/car_price.csv')
data = data.dropna()
print(data.head())
```

```python
def normal(x):
    maxi=max(x)
    mini=min(x)
    a=(x-mini)/(maxi-mini)
    return(a)
```

```python
y=data['selling_price']
year=normal(data['year'])
km=normal(data['km_driven'])
owner=data.owner.map({'First Owner':0, 'Second Owner':1,'Third Owner':2,'Fourth & Above Owner':3,'Test Drive Car':4},)
mileage = normal(pd.to_numeric(data['mileage'].map(lambda x: x.rstrip(' kmplkmkg/'))))
seats=data.seats.map({2:0,4:1,5:2,6:3,7:4,8:5,9:6,10:7,14:8})
```

```python
m=len(y)
x0=np.ones(m)
x=np.array([x0,year,km,owner,mileage,seats]).T
b=np.zeros(6)
y=np.array(y)
```

```python
def costfunction(x,y,b):
    m=len(y)
    ypred=x.dot(b)
    cost=np.sum((ypred-y))**2/(2*m)
    return(cost)
```

```python
def gradesc(x,y,b):
    time=120000
    a=0.05
    costhist=np.zeros(time)
    for i in range(time):
        h=x.dot(b)
        loss=h-y
        gradient=x.T.dot(loss)/m
        b=b-a*gradient
        cost=costfunction(x,y,b)
        costhist[i]=cost
    return(b,h,costhist)
```

```python
x=np.array([x0,year,km,owner,mileage,seats]).T
b=np.zeros(6)
a,b,c=gradesc(x,y,b)
```
